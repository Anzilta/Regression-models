```python
import numpy as np
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import fetch_california_housing
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor,GradientBoostingRegressor  
from sklearn.svm import SVR
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
```

```python
house=fetch_california_housing()
house_df=pd.DataFrame(house.data,columns=house.feature_names)
house_df['Target']=house.target
house_df.head(10)
```

```python
house_df.info()
```

```python
house_df.isnull().sum()
```

```python
scaler=StandardScaler()

Scaled_house=scaler.fit_transform(house_df)
scaled_df=pd.DataFrame(Scaled_house,columns=house_df.columns)

scaled_df.head(10)
```

<h3>Created a DataFrame:</h3><p>Converted the California Housing data into a table format (Pandas DataFrame) to make it much easier to read and work with.</p>

<h3>Checked for Missing Data:</h3><p>Looked for any blank and confirmed the dataset was already 100% complete with no missing values. Finding these early prevents errors and keeps the machine learning model accurate.</p>

<h3>Scaled the Data</h3><p>Used StandardScaler to put all the numbers onto a similar scale, making sure no single feature accidentally overpowers the others</p>

```python
x=scaled_df.drop("Target",axis=1)
y=scaled_df["Target"]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=42,test_size=0.2)
```

# Linear Regression Model

```python
lr=LinearRegression()
lr.fit(x_train,y_train)
```

```python
y_pred_lr=lr.predict(x_test)
```

```python
print("mse=",mean_squared_error(y_test,y_pred_lr))
print("mae=",mean_absolute_error(y_test,y_pred_lr))
print("R2_score=",r2_score(y_test,y_pred_lr))
```

# Decision Tree Regressor  

```python
dt=DecisionTreeRegressor(random_state=42)
dt.fit(x_train,y_train)
```

```python
y_pred_dt=dt.predict(x_test)
```

```python
print("mse=",mean_squared_error(y_test,y_pred_dt))
print("mae=",mean_absolute_error(y_test,y_pred_dt))
print("R2_score=",r2_score(y_test,y_pred_dt))
```

# Random Forest Regressor  

```python
rf=RandomForestRegressor(random_state=42,n_estimators=100)
rf.fit(x_train,y_train)
```

```python
y_pred_rf=rf.predict(x_test)
```

```python
print("mse=",mean_squared_error(y_test,y_pred_rf))
print("mae=",mean_absolute_error(y_test,y_pred_rf))
print("R2_score=",r2_score(y_test,y_pred_rf))
```

# Gradient Boosting Regressor  

```python
gb=GradientBoostingRegressor(random_state=42)
gb.fit(x_train,y_train)
```

```python
y_pred_gb=gb.predict(x_test)
```

```python
print("mse=",mean_squared_error(y_test,y_pred_gb))
print("mae=",mean_absolute_error(y_test,y_pred_gb))
print("R2_score=",r2_score(y_test,y_pred_gb))
```

# Support Vector Regressor (SVR)  

```python
svr=SVR()
svr.fit(x_train,y_train)
```

```python
y_pred_svr=svr.predict(x_test)
```

```python
print("mse=",mean_squared_error(y_test,y_pred_svr))
print("mae=",mean_absolute_error(y_test,y_pred_svr))
print("R2_score=",r2_score(y_test,y_pred_svr))
```

<h1>Conclusion</h1><p>After implementing all the regression models, the Random Forest Regressor gave the best performance with the highest R² score of 0.80. It predicted house prices more accurately because it can handle complex and non-linear data patterns effectively.

The Linear Regression model gave the lowest R² score. This is because it works mainly for linear relationships and could not capture the complex patterns in the California Housing dataset.

Therefore, Random Forest Regressor was the best model for this dataset, while Linear Regression was the least effective model.</p>
